```python
import pandas as pd

# Load hasil akhir kamu
df = pd.read_csv('dataset_final_super_lengkap.csv')

print("=== LAPORAN DETEKTIF DATA ===")

# 1. CEK CPU YANG MENCURIGAKAN (Skor Default 3000)
# Kenapa 3000? Karena di fungsi get_cpu_score, return paling bawah adalah 3000
cpu_misterius = df[df['CPU_Score'] == 3000]

if len(cpu_misterius) > 0:
    print(f"\n⚠️ Ditemukan {len(cpu_misterius)} Laptop dengan CPU TIDAK DIKENALI:")
    print(cpu_misterius['Processor'].unique()) # Tampilkan nama uniknya saja
else:
    print("\n✅ CPU Aman! Semua dikenali (atau setidaknya masuk logika tebakan i3/i5/i7).")

# 2. CEK GPU YANG MENCURIGAKAN (Skor Default 1500)
# Kenapa 1500? Karena di fungsi get_gpu_score_2025, return paling bawah adalah 1500
gpu_misterius = df[df['GPU_Score'] == 1500]

if len(gpu_misterius) > 0:
    print(f"\n⚠️ Ditemukan {len(gpu_misterius)} Laptop dengan GPU TIDAK DIKENALI:")
    print(gpu_misterius['GPU'].unique())
else:
    print("\n✅ GPU Aman! Semua dikenali.")
```
